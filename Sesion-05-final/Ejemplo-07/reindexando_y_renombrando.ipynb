## Ejemplo 7: Reindexando y renombrando columnas

### 1. Objetivos:
    - Limpiar un poco más nuestro dataset asignándole un índice y nombres de columnas apropiadas
 
---
    
### 2. Desarrollo:

Limpiemos nuestro dataset hasta que esté justo como lo dejamos en el Ejemplo pasado:

```python
import pandas as pd
```

```python
df = pd.read_csv('../../Datasets/melbourne_housing-raw.csv')
```

```python
df
```

```python
# quito las columnas ['BuildingArea', 'YearBuilt']
# como lo hicimos en el anterior jupyter notebook
df_2 = df.drop(columns=['BuildingArea', 'YearBuilt'])
```

```python
df_2
```

```python
# cambio los nans del Regionname por la palabra Unknown
df_2['Regionname'] = df_2['Regionname'].fillna('Unknown')
```

```python
df_2
```

```python
# Quito los renglones donde haya al menos un NAN y
# lo meto en el objeto df_dropped
df_dropped = df_2.dropna(axis=0, how='any')
```

```python
df_dropped
```

Ahora, tenemos dos situaciones:

1. La primera es que nuestro índice no coincide con el número de filas que tenemos.

```python
df_dropped
```

En este caso, dado que nuestro índice es secuencial y numérico, y no tiene ningún significado además de eso, nos convendría que reflejara la cantidad de filas que tenemos en nuestro dataset.

Para lograr eso vamos a usar el método `reset_index`:

```python
df_dropped.reset_index()
# el índice anterior lo vuelve columna
# reinicia en 0,1,2,3,,,, al índice nuevo
```

Nuestro índice ya está correcto, pero ahora tenemos un columna llamada `index` que contiene el índice original. Como no queremos guardar esos datos, agregamos la opción `drop=True` para eliminar el índice anterior:

```python
df_dropped.reset_index(drop=True)
# con drop = True, nos evitamos que cree una nueva columna
# llamada index
```

Guardemos nuestros cambios:

```python
# Lo asignamos a un objeto
df_dropped = df_dropped.reset_index(drop=True)
```

```python
df_dropped
```

Ahora tenemos un problema con los nombres de las columnas: Tienen inconsistencias en la manera cómo están nombradas y algunas incluso tienen errores ortográficos. Vamos a cambiarles los nombres para tener consistencia:

```python
# creamos un diccionario con el nombre viejo y el nombre nuevo
column_name_mapping = {
    'Suburb': 'suburb',
    'Address': 'address',
    'Rooms': 'rooms',
    'Type': 'type',
    'Price': 'price',
    'Method': 'method',
    'SellerG': 'seller_g',
    'Date': 'date',
    'Distance': 'distance',
    'Postcode': 'post_code',
    'Bedroom2': 'bedrooms',
    'Bathroom': 'bathroom',
    'Car': 'car',
    'Landsize': 'land_size',
    'CouncilArea': 'council_area',
    'Lattitude': 'latitude',
    'Longtitude': 'longitude',
    'Regionname': 'region_name',
    'Propertycount': 'property_count'
}
```

```python
# ahora voy a renombrar las columnas
# con los nombres correctos
# ocupo la función rename
df_renamed = df_dropped.rename(columns=column_name_mapping)

df_renamed
```

¡Listo! Nuestro dataset va agarrando forma.
